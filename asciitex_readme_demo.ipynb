# AsciiTeX README-style notebook

This notebook is a **self-contained example** that can live directly in the GitHub repository.

It introduces **AsciiTeX** as a small TeX-like compiler for **monospaced text layouts**.
The goal is to make the notebook work as a readable README companion:

- it explains what AsciiTeX does
- it shows the overall workflow
- it demonstrates sections, references, equations, diagrams, image inclusion, and bibliography
- it writes the tiny demo assets it needs by itself


## Repository assumptions

The notebook assumes that the repository contains these Python files next to it:

- `asciitex.py`
- `asciitex_math_extension.py`
- `asciitex_image_extension.py`
- `asciitex_bib_extension.py`
- `asciitex_diagram_extension.py`
- `asciitex_layout_extension.py`

The setup cell below creates:

- `README_demo.bib`

That keeps the example portable and easy to copy into a public repository.


```python

from pathlib import Path
import sys

# Make local repository files importable when the notebook is opened from the repo root.
repo_root = Path.cwd()
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

bib_text = r"""@article{knuth1981,
  author  = {Donald E. Knuth and Michael F. Plass},
  title   = {Breaking Paragraphs into Lines},
  journal = {Software: Practice and Experience},
  year    = {1981},
  volume  = {11},
  pages   = {1119--1184}
}

@book{lamport1994,
  author    = {Leslie Lamport},
  title     = {LaTeX: A Document Preparation System},
  publisher = {Addison-Wesley},
  year      = {1994}
}

@book{bringhurst2004,
  author    = {Robert Bringhurst},
  title     = {The Elements of Typographic Style},
  publisher = {Hartley and Marks},
  year      = {2004}
}
"""
o=Path("README_demo.bib").write_text(bib_text, encoding="utf-8")
```

## Compile one README-style AsciiTeX document

The next cell generates a compact document that explains **what AsciiTeX is**.

It is intentionally written like a one-page README example rather than a research note.


```python

from asciitex import TexLikeMonospaceCompiler, ExtensionRegistry
from asciitex_bib_extension import BibCiteExtension
from asciitex_math_extension import AsciiMathEquationExtension
from asciitex_image_extension import AsciiIncludeImageExtension
from asciitex_diagram_extension import DiagramPlotExtension
from asciitex_layout_extension import LayoutBlocksExtension


def build_readme_demo() -> str:
    return r"""
\header{AsciiTeX — TeX-like documents for monospaced output}
\footer{README-style demo}
\title{Typesetting monospaced documents in python using AsciiTeX}

\quote{AsciiTeX is a small TeX-like compiler for plain-text layouts.
It is useful when a document should stay readable as source,
render cleanly in a terminal, and still support structure,
references, equations, diagrams, and bibliography entries.
}

\hr

\section{What AsciiTeX is}
\label{sec:what}

AsciiTeX turns structured source into a monospaced document.
The syntax is intentionally familiar to anyone who has seen TeX-like markup.

A typical workflow is simple.

\begin{enumerate}
\item Write a source string or text file.
\item Register the extensions you need.
\item Compile the document to fixed-width output.
\end{enumerate}

This document itself demonstrates sections, references, display equations,
a generated diagram, an included image, a two-column block, and a bibliography.
See Section \ref{sec:features}, Equation \ref{eq:cost}, Diagram \ref{dia:pipeline},
and Figure \ref{fig:asset}.

\label{sec:features}
\section{Core ideas}

AsciiTeX focuses on text-first technical documents.

\begin{itemize}
\item readable source
\item monospaced rendering
\item automatic numbering for sections, figures, diagrams, and equations
\item labels and cross references
\item layout helpers for quotes, lists, rules, headers, and footers
\item optional extensions for math, diagrams, and richer layout blocks
\end{itemize}

A simple paragraph scoring function can be used to explain why line breaking
is a natural example for AsciiTeX.

\label{eq:cost}
\begin{equation}
C = \sum_{i=1}^{m} p_i
\end{equation}

Equation \ref{eq:cost} is intentionally simple. The point here is not a full
typesetting theory but a compact demonstration of numbered display equations.

\label{dia:pipeline}
\begindiagram[width=\columnwidth,height=12,place=h,mode=python,caption="Compilation pipeline",frame=true]
xs = [0, 1, 2, 3]
ys = [1, 2, 3, 4]

__diagram__ = {
    "type": "lines",
    "title": "Compilation pipeline",
    "x_label": "stage",
    "y_label": "progress",
    "grid": True,
    "legend": True,
    "lines": [
        {
            "x": xs,
            "y": ys,
            "name": "pipeline",
            "ch": "•",
            "dashed": False
        }
    ]
}
\enddiagram

Diagram \ref{dia:pipeline} summarizes the high-level idea: structured source goes
through parsing, numbering, and layout before plain-text output is produced.

\section{Included asset}
\label{sec:image}

AsciiTeX also supports an image inclusion command.

\label{fig:asset}
\includeimage[width=\columnwidth,height=8,place=h,palette=basic,invert=true,frame=true]{image.png}

\section{Two-column block}
\label{sec:twocol}

\begin{twocolumns}[textwidth=\textwidth,gutter=4,balance=true]

\subsection{Why it is useful}

AsciiTeX is a good fit for repositories that want examples which are:
easy to diff, easy to inspect in raw form, and easy to render in a terminal.

\begin{itemize}
\item no binary document format
\item source stays compact
\item generated output stays text-based
\end{itemize}

\subsection{Typical use cases}

Typical use cases include tutorials, technical notes, CLI demos, test fixtures,
and reference documents that should remain readable without a GUI.


\subsection{A tiny source fragment}

\verbatim{
  \section{Example}
  A small paragraph.
  
  \begin{equation}
    a^2 + b^2 = c^2
  \end{equation}
}

\columnbreak
\subsection{Related work}
Classic references for line breaking and document preparation include
\cite{knuth1981}, \cite{lamport1994}, and \cite{bringhurst2004}.

\section{Bibliography}
\label{sec:bib}

The bibliography below is loaded from an external file.

\bibliography{README_demo.bib}

\section{License}
Asciitex is released under the MIT License. 
\end{twocolumns}
"""


registry = ExtensionRegistry()
registry.add(BibCiteExtension(default_style="unsrt"))    
registry.add(AsciiMathEquationExtension())
registry.add(DiagramPlotExtension())
registry.add(AsciiIncludeImageExtension())
registry.add(LayoutBlocksExtension())

compiler = TexLikeMonospaceCompiler(registry=registry)

with open("third_party/hyphenation/hyph-en-us.pat.txt", "r", encoding="utf-8") as f:
    compiler.typesetter.load_hyphenation_patterns_text(f.read())
    
rendered = compiler.compile(
    build_readme_demo(),
    canvas_width=110,
    canvas_height=None,
    margin_left=2,
    margin_right=2,
    margin_top=1,
    margin_bottom=1,
    line_gap=1,
)

print(rendered)
```
